### Importing Necessary Libraries

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

### Tokenizing 

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([sentences])
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
len(tokenizer.word_index)
```

### Making Input Sequences

```python
input_sequences = []
for sentence in sentences.split('?'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
input_sequences[:10]
```

```python
max_len = max([len(x) for x in input_sequences])
max_len
```

### Padding the input sequence to have fixed Length

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
X_len=X.shape[1]
X_len
```

```python
y.shape
```

```python
len(tokenizer.word_index)
```

### Converting all unique words as an output category

```python
#we will add 1 bcz word_index starts from 1.and ohe always starts from 0 so it will miss the last word.
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=700)
```

```python
y.shape
```

```python
# Another necessary Imports
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense,Bidirectional,GRU
```

### Architechture of the Model

```python
model = Sequential()
model.add(Embedding(700,120, input_length=X_len))
model.add((LSTM(units=200, return_sequences=True)))
model.add((LSTM(units=200, return_sequences=True)))
model.add((LSTM(units=200, return_sequences=True)))
model.add((GRU(units=200)))
model.add(Dense(700, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'],)
```

### Summary of the Model

```python
model.summary()
```

### Training the model

```python
model.fit(X,y,epochs=150)
```

### Testing of the trained model

```python
import numpy as np

# Get text input from user
text = input("Enter the initial text: ")

# Get the length of range from user
range_length = int(input("How many next words you want to predict? "))

for i in range(range_length):
    # tokenize
    token_text = tokenizer.texts_to_sequences([text])[0]
    # padding
    padded_token_text = pad_sequences([token_text], maxlen=X_len, padding='pre')
    # predict
    pos = np.argmax(model.predict(padded_token_text, verbose=0))

    for word, index in tokenizer.word_index.items():
        if index == pos:
            text = text + " " + word

# Print the line in bold
print("\033[1mNext Predicted word/words:\033[0m", end='\n')
print(text)
```
